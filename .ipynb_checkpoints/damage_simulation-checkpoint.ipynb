```python


def _compute_combat_damage(unit1_strength, unit2_strength):
    diff = unit1_strength - unit2_strength

    return 30 * np.exp(diff / 25 * np.random.uniform(0.75, 1.25, size=diff.shape))


x = np.linspace(0, 60, 1000)
plt.scatter(x, _compute_combat_damage(x, 0), s=2, alpha=0.5)
plt.show()
```
