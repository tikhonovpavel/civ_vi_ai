```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

# mpl.rcParams['axes.facecolor'] = 'FFFFFF'

# plt.rcParams.update({'figure.facecolor': 'blue'})
# plt.rcParams.update({'axes.facecolor':'black'})
# plt.axes.set_facecolor(color=None)
```

```python
def _compute_combat_damage1(unit1_strength, unit2_strength):
    diff = unit1_strength - unit2_strength    
    return 30 * np.exp(diff / 25 * np.random.uniform(0.75, 1.25, size=diff.shape))



def _compute_combat_damage2(unit1_strength, unit2_strength):
    diff = unit1_strength - unit2_strength
    return np.random.uniform(0.8, 1.2, size=diff.shape) * 30 * np.exp(diff / 25)





x = np.linspace(-70, 40, 10000)
plt.scatter(x, _compute_combat_damage1(x, 0), s=2, alpha=0.2)
plt.axhline([100], c='red')
plt.xticks(np.arange(min(x), max(x)+1, step=10))
plt.grid(linewidth=0.2)
plt.show()

plt.scatter(x, _compute_combat_damage2(x, 0), s=2, alpha=0.2)
plt.axhline([100], c='red')
plt.xticks(np.arange(min(x), max(x)+1, step=10))
plt.grid(linewidth=0.2)
plt.show()
```
